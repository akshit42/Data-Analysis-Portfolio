```python
import pandas as pd
# Importing the python pandas library
```

```python
fifa = pd.read_csv("fifa21 raw data v2.csv", low_memory = False)
# Importing the dataset which will be worked on
```

```python
fifa.head(10)
# Getting a look at top 10 rows of the dataset
```

```python
fifa.describe
```

```python
fifa["Club"] = fifa["Club"].str.strip()
# Cleaning the Club column
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows', 20)
# Changing Jupyter Notebook settings
```

```python
fifa["Club"].dtype
```

```python
fifa["Height"] = fifa["Height"].str.strip('cm')
# CLeaning the Height column
```

```python
fifa["Weight"] = fifa["Weight"].str.strip('kg')
```

```python
fifa.head()
# Cleaning the Weight column
```

```python
fifa["Value"] = fifa["Value"].str.strip('€')
fifa["Wage"] = fifa["Wage"].str.strip('€')
fifa["Release Clause"] = fifa["Release Clause"].str.strip('€')
# Removing the € symbols from Value, Wage and Release Clause columns
```

```python
fifa["SM"] = fifa["SM"].str.strip('★')
fifa["IR"] = fifa["IR"].str.strip('★')
fifa["W/F"] = fifa["W/F"].str.strip('★')
# Cleaning the SM, IR, W/F columns of the ★ symbol
```

```python
def repl(x):
    if "M" in x:
        y = x.strip("M")
        return float(y)*1000000
    elif "K" in x:
        z = x.strip("K")
        return float(z)*1000
```

```python
fifa["Value"] = fifa["Value"].apply(repl)
# Removing M and K from Value column and converting the values to x1000000 and x1000 float values
```

```python
fifa["Wage"] = fifa["Wage"].apply(repl)
# Removing M & K from Wage column and converting the values to x1000000 and x1000 float values
```

```python
fifa["Release Clause"] = fifa["Release Clause"].apply(repl)
# Removing M & K from Release Clause column and converting the values to x1000000 and x1000 float values
```

```python
fifa
```

```python
fifa.to_csv("Fifa21_Clean.csv")
# Exporting the changes to a new csv file
```

```python
fifa["Joined"] = pd.to_datetime(fifa["Joined"])
# Converting Joined column values to pandas Timestamp type
```

```python
fifa.to_csv("Fifa21_Clean.csv")
# Saving the changes to the csv file
```

```python
def tenplus (t):
    return pd.to_datetime(t) + pd.DateOffset(years = 10)
# Creating a custom function to apply to Joined column
```

```python
fifa["Joined Plus Ten"] = fifa["Joined"].apply(tenplus)
# Simply adding 10 years to the date of joining of players
```

```python
fifa["Joined Plus Ten"]
```

```python
today = pd.to_datetime("today")
```

```python
today
```

```python
def more10 (m):
    return today > m
```

```python
fifa["Joined More Than Ten"] = fifa["Joined Plus Ten"].apply(more10)
# Comparing the date of this Jupyer Notebook's code compilation to Joined Plus Ten placeholder column
# This allows to find out players who have spent more than 10 years at their current club
# The results are store in Joined More Than Ten column
```

```python
fifa.drop(columns = "Joined Plus Ten", inplace = True)
#The placeholder Joined Plus Ten column is dropped
```
